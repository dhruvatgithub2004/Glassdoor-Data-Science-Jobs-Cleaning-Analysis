```python
def process_value(value):
    value = value.replace('$signed', '').replace('(USD)', '').strip()
    if 'to' in value:
        value = value.split('to')[1]
    value = value.replace('+', '').replace('$', '').strip()

    return value

df['Column2'] = df['Revenue'].apply(process_value)
```

## Data Analysis

### Which industries offer the highest salary estimates for data scientists?

```python
grouped = df.groupby("Industry")["Salary Estimate"].mean().sort_values(ascending=False)
pd.set_option('display.max_rows', None)
print(grouped)
pd.reset_option('display.max_rows')
```

### Which Sector Provides Highest salary Estimates to Data Scientist ?

```python
df.groupby("Sector")["Salary Estimate"].mean().sort_values(ascending=False)
```

### What Type of Ownership Companies Provide Better Salary Estimates ?

```python
df.groupby("Type of ownership")["Salary Estimate"].mean().sort_values(ascending=False).sort_values(ascending=False)
```

```python
sns.barplot(data = df.groupby("Type of ownership")["Salary Estimate"].mean().sort_values(ascending=False).sort_values(ascending=False))
plt.xticks(rotation=90)  # Rotate x-axis labels by 90 degrees
plt.xlabel('Type of Ownership')
plt.ylabel('Salary Estimate Sum')
plt.title('Salary Estimate Sum by Industry')
plt.tight_layout()
plt.show()
```

### What Size company provides Better Salary Estimates ?

```python
df.groupby("Size")["Salary Estimate"].mean().sort_values(ascending=False)
```

### Which Cities and States has better Rated Companies ?

```python
 df.groupby(["City","State"])["Rating"].mean().sort_values(ascending=False).head()
```

### Are there any correlations between the company ratings and the salary estimates?
